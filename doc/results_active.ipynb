```python
dt = d1[['ID', 'Sequence', 'Caloric Intake', 'Height', 'Weight', 'Godin_PA', 'BMI', 'LBM (kg)', 'Bone_Age', 
         'Tanner', 'BF_Slaughter', 'PAQ_TOTAL_SCORE', 'PAQ-PRVNT', 'tot hard', 'Vit D', 'Calcium']]
DM = pandas.merge(d1, d2, left_on=['ID', 'Sequence'], right_on=['id', 'sequence'], how='right'); del dt
DM.head()
```

```python
dm = DM.copy()
dm = dm.sort_values(by=['id', 'sequence'])

dm['boy']  = numpy.where(dm.gender=="boy", 1, 0)
dm['girl'] = numpy.where(dm.gender=="girl", 1, 0)

dm['trip'] = dm.groupby(['id']).cumcount()+1
dm = dm.rename(columns={'sequence': 'session'})
dm = dm.rename(columns={'Biodex': 'biodex'})
dm = dm.rename(columns={'MatLab': 'matlab'})
dm = dm.rename(columns={'Caloric Intake': 'calo'})
dm = dm.rename(columns={'Height': 'height'})
dm = dm.rename(columns={'Weight': 'weight'})
dm = dm.rename(columns={'LBM (kg)': 'LBM'})
dm = dm.rename(columns={'tot hard': 'tot_hard'})
dm = dm.rename(columns={'PAQ-PRVNT': 'PAQ_PRVNT'})
dm = dm.rename(columns={'Vit D': 'vit_d'})


def get_st(var):
    temp = 'st_' + var
    dm[temp] = (dm[var] - dm[var].mean()) / dm[var].std()

_vars = ['rsos', 'tsos', 'grip', 'ptiso', 'biodex', 'matlab', 'ntxc', 'calo', 'matu', 'mvh', 'age', 'height', 'weight', 
         'Godin_PA', 'BMI', 'LBM', 'PAQ_TOTAL_SCORE', 'PAQ_PRVNT', 'BF_Slaughter', 'tot_hard', 'vit_d', 'Caffeine', 'Bone_Age', 'Tanner']
for v in _vars:
    get_st(v)

#dm.to_csv('out/data_analysis.csv', index=False)
dm.head()
```
