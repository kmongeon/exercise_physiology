```python
import pandas
import numpy
import ipystata
from ipystata.config import config_stata
config_stata('/usr/local/stata/stata-mp')

pandas.set_option('display.float_format', lambda x: '%.4f' % x)
pandas.set_option('precision', 4)
pandas.set_option('max_columns', 400)
pandas.set_option('mode.sim_interactive', True)
pandas.set_option('colheader_justify', 'center')

from IPython.display import HTML, Markdown, Latex, Image
```

```python
d1 = pandas.read_csv('./_data/IzzyLongitudinalWorkingMarch2015.csv')
d2 = pandas.read_csv('./_data/AnalysisObservations.csv')
dt = pandas.merge(d1, d2, left_on=['ID', 'Sequence'], right_on=['id', 'sequence'], how='right')
del d1, d2
```

```python
dm = dt.copy()
dm = dm.sort_values(by=['id', 'sequence'])

year = []
for row in dt.Sequence:
    if row==1:
       year.append(2010)
    elif row==2:
       year.append(2010)
    elif row==3:
       year.append(2011)
    elif row==4:
       year.append(2011)
    elif row==5:
       year.append(2012)
    elif row==6:
       year.append(2012)
    elif row==7:
       year.append(2013)
    else:
       year.append('na')
season = []
for row in dt.Sequence:
    if row==1:
       season.append('fall')
    elif row==2:
       season.append('spring')
    elif row==3:
       season.append('fall')
    elif row==4:
       season.append('spring')
    elif row==5:
       season.append('fall')
    elif row==6:
       season.append('spring')
    elif row==7:
       season.append('fall')
    else:
       season.append('na')
dm['year'] = year
dm['season'] = season

dm['trip'] = dm.groupby(['id']).cumcount()+1
dm['boy']  = numpy.where(dm.gender=="boy", 1, 0)
dm['girl'] = numpy.where(dm.gender=="girl", 1, 0)

dm = dm.rename(columns={'sequence': 'session'})
dm = dm.rename(columns={'Biodex': 'biodex'})
dm = dm.rename(columns={'MatLab': 'matlab'})
dm = dm.rename(columns={'Caloric Intake': 'calo'})
dm = dm.rename(columns={'Height': 'height'})
dm = dm.rename(columns={'Weight': 'weight'})
dm = dm.rename(columns={'LBM (kg)': 'LBM'})
dm = dm.rename(columns={'tot hard': 'tot_hard'})
dm = dm.rename(columns={'PAQ-PRVNT': 'PAQ_PRVNT'})
dm = dm.rename(columns={'Vit D': 'vit_d'})
dm = dm.rename(columns={'NTx/Creat': 'NTx_Creat'})
dm = dm.rename(columns={'Quads.1': 'Quads_1'})
    
def get_st(var):
    temp = 'st_' + var
    dm[temp] = (dm[var] - dm[var].mean()) / dm[var].std()

_vars = ['rsos', 'tsos', 'grip', 'ptiso', 'biodex', 'matlab', 'ntxc', 'calo', 'matu', 'mvh', 'age', 'height', 'weight', 
         'Godin_PA', 'BMI', 'LBM', 'PAQ_TOTAL_SCORE', 'PAQ_PRVNT', 'BF_Slaughter', 'tot_hard', 'vit_d', 'Caffeine', 'Bone_Age', 'Tanner']
for v in _vars:
    get_st(v)

dm.columns = dm.columns.str.replace('\s+', '_')  
dm.columns = dm.columns.str.replace('.', '_') 

DM = dm.copy()
DM.to_csv('DM.csv', index=False)
DM.to_stata('DM.dta')

%store DM
```
